Необходимо найти все целые числа от 1 до 400 000, для которых sin от этого числа будет иметь не менее 9 одинаковых цифр (например, sin(139198)=0.30763333543133703; sin(139198)=0.30763333543133703 т.е. имеет 9 троек). Ответ представляет собой список пар (целое число, значение синуса для него), для которых значение синуса удовлетворяет условию. Решение этой задачи распараллелить, используя multiprocessing Pool. Сравнить продолжительность последовательного и параллельного решения задачи.

```python
#Последовательное решение
import math
import time

def check_sin(n):
    sin_n = math.sin(n)
    sin_str = str(sin_n).replace('.', '')
    for i in range(10):
        if str(i) * 9 in sin_str:
            return (n, sin_n)
    return None
# `check_sin` принимает целое число `n`, вычисляет его синус с помощью функции `sin` из модуля `math`,
# и проверяет, имеет ли значение синуса не менее 9 одинаковых цифр. Если значение синуса удовлетворяет условию,
# функция возвращает пару `(n, sin_n)`, иначе возвращает `None`.

start_time = time.time()

results = []
for n in range(1, 400001):
    result = check_sin(n)
    if result is not None:
        results.append(result)
# Запускаем цикл от 1 до 400000 и для каждого числа вызываем функцию `check_sin`.
# Если функция возвращает результат, мы добавляем его в список `results`.

end_time = time.time()

print(f'Found {len(results)} results in {end_time - start_time:.2f} seconds.')
print(results)
```

```python
import math
import multiprocessing
import time

def check_sin(n):
    sin_n = math.sin(n)
    sin_str = str(sin_n).replace('.', '')
    for i in range(10):
        if str(i) * 9 in sin_str:
            return (n, sin_n)
    return None

start_time = time.time()

with multiprocessing.Pool(processes=1) as pool:
    results = pool.map(check_sin, range(1, 400001))
# создаем пул процессов с помощью класса `Pool`. Затем мы используем метод `map` для вызова функции `check_sin` 
# для каждого числа от 1 до 400000. Метод `map` автоматически распределяет вычисления между процессами.
results = [result for result in results if result is not None]
# фильтруем результаты, удаляя `None`, и выводим количество найденных результатов и время выполнения программы
end_time = time.time()

print(f'Found {len(results)} results in {end_time - start_time:.2f} seconds.')
```

```python
import math
from joblib import Parallel, delayed
import numpy as np

# функция для проверки, имеет ли число sin с 9 одинаковыми цифрами
def check_sin(i):
    sin_i = math.sin(i)
    digits = [int(d) for d in str(abs(int(sin_i * 10**9)))]
    for j in range(10):
        if digits.count(j) >= 9:
            return (i, sin_i)
    return None

# параллельное решение с помощью joblib
results_par = Parallel(n_jobs=-1)(delayed(check_sin)(i) for i in range(1, 400001))
results_par = [res for res in results_par if res is not None]

# вывод результатов
print('Результаты:')
for res in results_par:
    print(res)
```

Напишите функцию на Python, которая вычисляет сумму квадратов всех чисел в списке размера N, содержащем целые случайно сгенерированные числа. Затем напишите функцию, скомпилированную с помощью Numba, которая выполняет ту же задачу, и сравните время выполнения обеих функций. Сравните скорость при различных N и сделайте вывод.

```python
import random
# определяем функцию `sum_of_squares_python`, которая принимает список `lst` и вычисляет сумму квадратов всех его элементов
# с помощью цикла.
def sum_of_squares_python(lst):
    result = 0
    for x in lst:
        result += x**2
    return result

# генерируем список случайных чисел
lst = [random.randint(0, 100) for _ in range(100000)]

# вычисляем сумму квадратов с помощью чистого Python
result = sum_of_squares_python(lst)

print(result)
```

```python
import random
import numba

# мы определяем функцию `sum_of_squares_numba` так же,
# как и в предыдущем примере, но добавляем декоратор `numba.jit` для компиляции функции в машинный код.
@numba.jit(nopython=True)
def sum_of_squares_numba(lst):
    result = 0
    for x in lst:
        result += x**2
    return result

# генерируем список случайных чисел
lst = [random.randint(0, 100) for _ in range(100000)]

# вычисляем сумму квадратов с помощью Numba
result = sum_of_squares_numba(lst)

print(result)
```

```python
import time

# генерируем список случайных чисел
lst = [random.randint(0, 100) for _ in range(100000)]

# засекаем время выполнения функции на чистом Python
start_time = time.time()
result_python = sum_of_squares_python(lst)
end_time = time.time()
print("Python time:", end_time - start_time)

# засекаем время выполнения функции с Numba
start_time = time.time()
result_numba = sum_of_squares_numba(lst)
end_time = time.time()
print("Numba time:", end_time - start_time)

# проверяем, что обе функции выдают одинаковый результат
assert result_python == result_numba
```

 Необходимо найти все пары целых чисел от 1 до 3000, для которых sin от произведения чисел из пары будет иметь значение больше 0.99999999 (например, sin(237∗1544)=0.9999999951152019>0.99999999; sin(237∗1544)=0.9999999951152019>0.99999999 ). Ответ представляет собой список пар (целое число1, целое число2, значение синуса для из произведения), для которых значение синуса удовлетворяет условию. Решение этой задачи распараллелить, используя multiprocessing Pool. Сравнить продолжительность последовательного и параллельного решения задачи.

```python
import math
import time

# засекаем время выполнения
start_time = time.time()

# проходим по всем парам чисел от 1 до 3000
results = []
for i in range(1, 3001):
    for j in range(1, 3001):
        # вычисляем произведение и синус
        prod = i * j
        sin_prod = math.sin(prod)
        # проверяем условие
        if sin_prod > 0.99999999:
            results.append((i, j, sin_prod))

# выводим результат и время выполнения
print("Results:", results)
print("Time:", time.time() - start_time)
```

```python
import math
import time
from multiprocessing import Pool

# определяем функцию для вычисления синуса произведения пары чисел
def sin_prod(pair):
    i, j = pair
    prod = i * j
    sin_prod = math.sin(prod)
    return (i, j, sin_prod)

# засекаем время выполнения
start_time = time.time()

# создаем список всех возможных пар чисел от 1 до 3000
pairs = [(i, j) for i in range(1, 3001) for j in range(1, 3001)]

# создаем пул процессов
with Pool() as pool:
    # распределяем вычисления по нескольким процессам
    results = pool.map(sin_prod, pairs)
    # фильтруем результаты по условию
    results = [result for result in results if result[2] > 0.99999999]

# выводим результат и время выполнения
print("Results:", results)
print("Time:", time.time() - start_time)
```

```python
import sqlite3
import pandas as pd
con = sqlite3.connect('Chinook_sqlite.sqlite') # Открываем базу данных 
cur = con.cursor() # Создаем объект-курсор
def all_artists(cur):
    query_1 = '''
    select *
    from artist
    '''
    cur.execute(query_1) # выполняем запрос
    fa = cur.fetchall() # возвращает список всех записей из результата запроса
    colnames = [i[0] for i in cur.description] #cursor.description предоставляет имена столбцов последнего запроса
    result = pd.DataFrame(fa,columns= colnames)
    return result

all_artists(cur)
```

 Подсчитать, сколько раз в текстовых файлах, лежащих в all_k.zip, встречаются предложения трех видов: вопросительные (в окончании имеют вопросительный знак), побудительные (в окончании имеют восклицательный знак и не имеют вопросительного) и повествовательные (в окончании имеют точку или троеточие, при этом нужно исключить учет точек, встречающихся в сокращениях, таких как ""т.к."").
Выполнить задание с использованием Dask (корректным!), распараллелив процесс обработки данных (использование Dask должно приводить к истинной параллельной обработке данных).


```python
import dask.bag as db
import re

# функция для подсчета количества вопросительных, побудительных и повествовательных предложений
def count_sentences(text):
    sentences = re.findall(r'\b\w[^.?!]*[.?!]', text)
    question_count = len([s for s in sentences if s.endswith('?')])
    exclamation_count = len([s for s in sentences if s.endswith('!') and not s.endswith('?')])
    narrative_count = len([s for s in sentences if s.endswith('.') or s.endswith('...')])
    return (question_count, exclamation_count, narrative_count)

# создаем dask.bag из всех текстовых файлов в папке all_k
bag = db.read_text('./data/data/all_k/*.txt', encoding = 'windows-1251')

# применяем функцию count_sentences к каждому файлу и получаем список кортежей с результатами
results = bag.map(count_sentences).compute()

# суммируем результаты для всех файлов
total_questions = sum([r[0] for r in results])
total_exclamations = sum([r[1] for r in results])
total_narratives = sum([r[2] for r in results])

# выводим результаты
print('Количество вопросительных предложений:', total_questions)
print('Количество побудительных предложений:', total_exclamations)
print('Количество повествовательных предложений:', total_narratives)
```

```python
import dask.dataframe as dd

# Читаем все файлы в директории, которые соответствуют шаблону
accounts = dd.read_csv("./data/data/accounts/accounts.*.csv", 
                       dtype={"amount": "float64", "id": "int64", "names": "object"})

# Отфильтровываем данные, оставляя только значения, кратные трем
multiples_of_three = accounts[accounts.amount % 3 == 0]

# Группируем данные по ID и считаем количество таких значений
grouped = multiples_of_three.groupby('id').size()

# Ищем ID с наибольшим количеством значений, кратных трем
result_id = grouped.idxmax().compute()

print(f"id, для которого в столбце amount встречается наибольшее количество значений, кратных трем {result_id}")

```

 В accounts.*.csv найти id, для которого сумма положительных значений в столбце amount наибольшая.

```python
import dask.dataframe as dd

# Читаем все файлы в директории, которые соответствуют шаблону
accounts = dd.read_csv("./data/data/accounts/accounts.*.csv", 
                       dtype={"amount": "float64", "id": "int64", "names": "object"})

# Отфильтровываем данные, оставляя только положительные значения
positive_amounts = accounts[accounts.amount > 0]

# Группируем данные по ID и суммируем amount
grouped = positive_amounts.groupby('id').amount.sum()

# Ищем ID с наибольшей суммой
result_id = grouped.idxmax().compute()

print(f" ID с наибольшей суммой положительных значений в столбце amount {result_id}")
```
